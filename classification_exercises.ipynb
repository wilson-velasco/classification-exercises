# Data Preparation

## Using the Iris Data:

```python
# 1. Use the function defined in acquire.py to load the iris data.

import acquire as a

iris = a.get_iris_data()
iris
```

```python
# 2. Drop the species_id and measurement_id columns.

iris = iris.drop(columns=['species_id', 'measurement_id'])

iris
```

```python
# 3. Rename the species_name column to just species.

iris = iris.rename(columns={'species_name':'species'})

iris
```

```python
# 4. Create dummy variables of the species name and concatenate onto the iris dataframe. (This is for practice, 
# we don't always have to encode the target, but if we used species as a feature, we would need to encode it).

iris_dummy = pd.get_dummies(iris.species, drop_first=True)

iris_dummy
```

```python
iris = pd.concat([iris, iris_dummy], axis=1)

iris
```

```python
# 5. Create a function named prep_iris that accepts the untransformed iris data, and returns the data with the 
# transformations above applied.

iris = a.prep_iris(iris)
```

```python
iris
```

## Using the Titanic dataset:

```python
# 1. Use the function defined in acquire.py to load the Titanic data.

titanic = a.get_titanic_data()

titanic
```

```python
# 2. Drop any unnecessary, unhelpful, or duplicated columns.

titanic = titanic.set_index('passenger_id').drop(columns=['class', 'embark_town', 'deck'])

titanic
```

```python
# 3. Encode the categorical columns. Create dummy variables of the categorical columns and concatenate them onto 
# the dataframe.

titanic_dummy = pd.get_dummies(titanic[['sex', 'embarked']], drop_first=[True,True])

pd.concat([titanic, titanic_dummy], axis=1)
```

```python
# 4. Create a function named prep_titanic that accepts the raw titanic data, and returns the data with the 
# transformations above applied.

titanic = a.prep_titanic(titanic)

titanic
```

## Using the Telco dataset

```python
# 1. Use the function defined in acquire.py to load the Telco data.

telco = a.get_telco_data()

telco
```

```python
# 2. Drop any unnecessary, unhelpful, or duplicated columns. This could mean dropping foreign key columns but 
# keeping the corresponding string values, for example.

telco.isnull().sum() # no null values

telco = telco.set_index('customer_id').drop(columns=['payment_type_id', 'internet_service_type_id', 'contract_type_id', 'paperless_billing'])

telco = telco.replace(['Yes', 'No'], [1,0])

telco
```

```python
telco.nunique() # just seeing who needs encoding, i.e. those between 3-4 unique values. those with 2 I 
```

```python
# 3. Encode the categorical columns. Create dummy variables of the categorical columns and concatenate them onto 
# the dataframe.

telco_dummies = pd.get_dummies(telco[['gender','multiple_lines', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies', 'contract_type', 'internet_service_type', 'payment_type']], drop_first=True)
```

```python
telco = pd.concat([telco, telco_dummies], axis=1)
telco
```

```python
# 4. Create a function named prep_telco that accepts the raw telco data, and returns the data with the 
# transformations above applied.

telco = a.prep_telco(telco)

telco
```
