```python
# For virginica & versicolor: Compare the mean petal_width using the Mann-Whitney test (scipy.stats.mannwhitneyu) 
# to see if there is a significant difference between the two groups. Do the same for the other measurement types.

from scipy import stats
```

$H_0$: There is no significant difference in the petal widths between virginica and versicolor
    
$H_\alpha$: There is a significant difference in the petal widths between virginica and versicolor
    
$\alpha$ = .05

```python
stats.mannwhitneyu(train_iris.petal_width[train_iris.species == 'virginica'], 
                   train_iris.petal_width[train_iris.species == 'versicolor'])

# Because the p-value is below alpha, we reject the null hypothesis.
```

$H_0$: There is no significant difference in the petal lengths between virginica and versicolor
    
$H_\alpha$: There is a significant difference in the petal lengths between virginica and versicolor
    
$\alpha$ = .05

```python
stats.mannwhitneyu(train_iris.petal_length[train_iris.species == 'virginica'], 
                   train_iris.petal_length[train_iris.species == 'versicolor'])

# Because the p-value is below alpha, we reject the null hypothesis.
```

$H_0$: There is no significant difference in the sepal widths between virginica and versicolor
    
$H_\alpha$: There is a significant difference in the sepal widths between virginica and versicolor
    
$\alpha$ = .05

```python
stats.mannwhitneyu(train_iris.sepal_width[train_iris.species == 'virginica'], 
                   train_iris.sepal_width[train_iris.species == 'versicolor'])

# Because the p-value is above alpha, we fail to reject the null hypothesis.
```

$H_0$: There is no significant difference in the sepal lengths between virginica and versicolor
    
$H_\alpha$: There is a significant difference in the sepal lengths between virginica and versicolor
    
$\alpha$ = .05

```python
stats.mannwhitneyu(train_iris.sepal_length[train_iris.species == 'virginica'], 
                   train_iris.sepal_length[train_iris.species == 'versicolor'])

## Because the p-value is below alpha, we reject the null hypothesis.
```

```python
# Document takeaways & any actions.


# Of all the bivariate tests, sepal_width is the only variable that is UNABLE to predict whether a species is 
# virginica vs. versicolor. 
# Of all the measurement variables, petal_width is the one that visually has the greatest difference across all 
# three species.
```

```python
# Ask a specific question of the data, such as: is the sepal area signficantly different in virginica compared to 
# setosa? Answer the question through both a plot and using a mann-whitney or t-test. If you use a t-test, be sure 
# assumptions are met (independence, normality, equal variance).

# Is the sepal area significantly different in virginica compared to setosa?
```

$H_0$: There is no difference in sepal area between virginica and setosa.

$H_\alpha$: There is a difference in sepal area between virginica and setosa.

```python
# First, create a new variable called sepal_area.

train_iris['sepal_area'] = train_iris.sepal_length * train_iris.sepal_width
```

```python
# Then, check out distributions visually.

sns.histplot(data=train_iris.sepal_area[train_iris.species =='virginica'])
plt.title('Virginica')
plt.show()
sns.histplot(data=train_iris.sepal_area[train_iris.species == 'setosa'])
plt.title('Setosa')
plt.show()
train_iris.species.value_counts()

# They more or less follow a normal distribution. Will use a two-sample, two-tailed T-Test.
```

```python
# Check variances

stats.levene(train_iris.sepal_area[train_iris.species =='virginica'], 
             train_iris.sepal_area[train_iris.species =='setosa'])

# Not different enough. Will need to set equal_var to False.
```

```python
stats.ttest_ind(train_iris.sepal_area[train_iris.species =='virginica'], 
                   train_iris.sepal_area[train_iris.species =='setosa'],
               equal_var = False)

# Yes, we reject the null hypothesis since p-value is less than alpha. 
```

```python
# Document takeaways and any actions.

# Petal length and petal width are the biggest predictors of species.
# Sepal_length and sepal_width were used to create new column sepal_area. An ok predictor.
```
